```python
nav_df = pd.read_csv("../data/raw/02_nav_history.csv")

print(nav_df.shape)
print(nav_df.head())
print(nav_df.columns)
```

```python
# Convert date

nav_df["date"] = pd.to_datetime(nav_df["date"])

# Sort

nav_df = nav_df.sort_values(
    ["amfi_code", "date"]
)

# Remove duplicates

duplicates_before = nav_df.duplicated().sum()

nav_df = nav_df.drop_duplicates()

duplicates_after = nav_df.duplicated().sum()

print("Duplicates before:", duplicates_before)
print("Duplicates after:", duplicates_after)
```

```python
invalid_nav = nav_df[
    nav_df["nav"] <= 0
]

print(
    "Invalid NAV rows:",
    len(invalid_nav)
)
```

```python
nav_df.to_csv(
    "../data/processed/clean_nav_history.csv",
    index=False
)

print("Saved successfully")
```

```python
import pandas as pd
from pathlib import Path

raw_dir = Path("../data/raw")
processed_dir = Path("../data/processed")

# 1. Fund Master
fund_master = pd.read_csv(raw_dir / "01_fund_master.csv")
fund_master.to_csv(
    processed_dir / "clean_fund_master.csv",
    index=False
)

# 2. AUM By Fund House
aum_df = pd.read_csv(raw_dir / "03_aum_by_fund_house.csv")

aum_df["date"] = pd.to_datetime(aum_df["date"])

aum_df.to_csv(
    processed_dir / "clean_aum_by_fund_house.csv",
    index=False
)

# 3. Monthly SIP Inflows
sip_df = pd.read_csv(raw_dir / "04_monthly_sip_inflows.csv")

sip_df["month"] = pd.to_datetime(
    sip_df["month"]
)

sip_df.to_csv(
    processed_dir / "clean_monthly_sip_inflows.csv",
    index=False
)

# 4. Category Inflows
cat_df = pd.read_csv(raw_dir / "05_category_inflows.csv")

cat_df["month"] = pd.to_datetime(
    cat_df["month"]
)

cat_df.to_csv(
    processed_dir / "clean_category_inflows.csv",
    index=False
)

# 5. Industry Folio Count
folio_df = pd.read_csv(
    raw_dir / "06_industry_folio_count.csv"
)

folio_df["month"] = pd.to_datetime(
    folio_df["month"]
)

folio_df.to_csv(
    processed_dir / "clean_industry_folio_count.csv",
    index=False
)

# 6. Portfolio Holdings
portfolio_df = pd.read_csv(
    raw_dir / "09_portfolio_holdings.csv"
)

if "portfolio_date" in portfolio_df.columns:
    portfolio_df["portfolio_date"] = pd.to_datetime(
        portfolio_df["portfolio_date"]
    )

portfolio_df.to_csv(
    processed_dir / "clean_portfolio_holdings.csv",
    index=False
)

# 7. Benchmark Indices
benchmark_df = pd.read_csv(
    raw_dir / "10_benchmark_indices.csv"
)

benchmark_df["date"] = pd.to_datetime(
    benchmark_df["date"]
)

benchmark_df.to_csv(
    processed_dir / "clean_benchmark_indices.csv",
    index=False
)

print("All remaining datasets cleaned and saved successfully.")
```

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(
    "sqlite:///../data/db/bluestock_mf.db"
)

tables = {
    "dim_fund":
        "../data/processed/clean_fund_master.csv",

    "fact_nav":
        "../data/processed/clean_nav_history.csv",

    "fact_transactions":
        "../data/processed/clean_investor_transactions.csv",

    "fact_performance":
        "../data/processed/clean_scheme_performance.csv",

    "fact_aum":
        "../data/processed/clean_aum_by_fund_house.csv"
}

for table_name, file_path in tables.items():

    df = pd.read_csv(file_path)

    df.to_sql(
        table_name,
        engine,
        if_exists="replace",
        index=False
    )

    print(
        f"{table_name}: {len(df)} rows loaded"
    )
```

```python
tables = pd.read_sql(
    "SELECT name FROM sqlite_master WHERE type='table';",
    engine
)

print(tables)
```
